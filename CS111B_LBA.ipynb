```python
import matplotlib.pyplot as plt 

# Markov Matrix 
M = matrix(QQ,[
        [40/100,0,25/100,12/100,0,0],
        [42/100,50/100,0,0,0,0],
        [0,0,75/100,0,20/100,0],
        [0,25/100,0,40/100,0,60/100],
        [0,0,0,48/100,80/100,0],
        [18/100,25/100,0,0,0,40/100]])


# the first distribution
v1 = vector([3/10, 2/10, 1/10, 2/10, 1/10, 1/10])

# the second distribution
v2 = vector([5/10, 0, 1/10, 1/10, 3/10, 0])

# the third distribution
v3 =vector([4/10, 1/10, 1/10, 2/10, 1/10, 1/10])



def migration(M, k, now): # Simulate migrations through Markov Chain 
    migs = []
    for i in range(1,k+1):
        migs.append(vector(M^(i)*now))
    return migs

hour = 15
hours = list(range(hour))
scenarios = migration(M, hour, v1)

node1 = []
node2 = []
node3 = []
node4 = []
node5 = []
node6 = []

for i in scenarios:
    node1.append(i[0])
    node2.append(i[1])
    node3.append(i[2])
    node4.append(i[3])
    node5.append(i[4])
    node6.append(i[5])

# SCENARIO 1 
plt.title("Scenario 1")
plt.plot(hours, node1, label = 'Entrance')
plt.plot(hours, node2, label = 'Tickets')
plt.plot(hours, node6, label = 'Exhibit 1')
plt.plot(hours, node4, label = 'Exhibit 2')
plt.plot(hours, node5, label = 'Open-air Exhibit')
plt.plot(hours, node3, label = 'Sovenenir Shop')

plt.xlabel("per hour") 
plt.ylabel("population ratio")
plt.legend(loc = "upper left")
plt.show()

# SCENARIO 2 
scenarios = migration(M, hour, v2)

node1 = []
node2 = []
node3 = []
node4 = []
node5 = []
node6 = []
k = []

for i in scenarios:
    node1.append(i[0])
    node2.append(i[1])
    node3.append(i[2])
    node4.append(i[3])
    node5.append(i[4])
    node6.append(i[5])

for i in range(len(scenarios)):
    k.append(i)

plt.title("Scenario 2")
plt.plot(hours, node1, label = 'Entrance')
plt.plot(hours, node2, label = 'Tickets')
plt.plot(hours, node6, label = 'Exhibit 1')
plt.plot(hours, node4, label = 'Exhibit 2')
plt.plot(hours, node5, label = 'Open-air Exhibit')
plt.plot(hours, node3, label = 'Sovenenir Shop')

plt.xlabel("per hour") 
plt.ylabel("population ratio")
plt.legend(loc = 'upper left')
plt.show()

# SCENARIO 3
scenarios = migration(M, hour, v3)

node1 = []
node2 = []
node3 = []
node4 = []
node5 = []
node6 = []

for i in scenarios:
    node1.append(i[0])
    node2.append(i[1])
    node3.append(i[2])
    node4.append(i[3])
    node5.append(i[4])
    node6.append(i[5])

plt.title("Scenario 3")
plt.plot(hours, node1, label = 'Entrance')
plt.plot(hours, node2, label = 'Tickets')
plt.plot(hours, node6, label = 'Exhibit 1')
plt.plot(hours, node4, label = 'Exhibit 2')
plt.plot(hours, node5, label = 'Open-air Exhibit')
plt.plot(hours, node3, label = 'Sovenenir Shop')

plt.xlabel("per hour") 
plt.ylabel("population ratio")
plt.legend(loc = 'upper left')
plt.show()



```

```python
### SCENARIO: Initial State Eigenvectors ### 

M = matrix(QQ, [[40/100,0,25/100,12/100,0,0],[42/100,50/100,0,0,0,0],[0,0,75/100,0,20/100,0],
                 [0,25/100,0,40/100,0,60/100],[0,0,0,48/100,80/100,0],[18/100,25/100,0,0,0,40/100]])

stable_vec = M.eigenvectors_right()[0][1][0]
stable_vec_norm = n(stable_vec/sum(stable_vec))

# Using Eigenvector
scenarios = migration(M, hour, stable_vec_norm)

node1 = []
node2 = []
node3 = []
node4 = []
node5 = []
node6 = []

for i in scenarios:
    node1.append(i[0])
    node2.append(i[1])
    node3.append(i[2])
    node4.append(i[3])
    node5.append(i[4])
    node6.append(i[5])
    
plt.title("Scenario: Starting with Eigenvector")
plt.plot(hours, node1, label = 'Entrance')
plt.plot(hours, node2, label = 'Tickets')
plt.plot(hours, node6, label = 'Exhibit 1')
plt.plot(hours, node4, label = 'Exhibit 2')
plt.plot(hours, node5, label = 'Open-air Exhibit')
plt.plot(hours, node3, label = 'Sovenenir Shop')

plt.xlabel("per hour") 
plt.ylabel("population ratio")
plt.legend(loc = 'upper left')
plt.show()
```
